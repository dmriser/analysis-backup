```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
data = pd.read_csv('xs.dat')
```

```python
xs = pd.read_csv('table.csv')
```

```python
data.dropna(how='any', inplace=True)
```

```python
mass_proton = 0.938
beam_energy = 5.5
alpha = 1/137.0

data.xs_keppel *= 2*np.pi

data['x'] = data.q2/(data.w**2 - mass_proton**2 + data.q2)
data['nu'] = data.q2/(2*mass_proton*data.x)
data['eprime'] = beam_energy - data.nu
data['y'] = 1-data.eprime/beam_energy
data['theta'] = 2*np.arcsin(np.sqrt(data.q2/(4*data.eprime*beam_energy)))
data['epsilon'] = 1/(1 + 2*(1+data.nu/data.q2)*np.tan(data.theta/2)**2)
data['photon_flux'] = alpha*data.eprime*(data.w**2-mass_proton**2)/(4*np.pi**2 *(data.q2*mass_proton*beam_energy*(1-data.epsilon)))
data['xs_bosted_corr'] = data.xs_bosted
#data['dw_dx'] = data.q2/(2*data.x**2*data.w)

data.head() 
```

```python
FONT_SIZE = 18

def direct_compare(data, xs=None):

    q2_bins = np.unique(data.q2_bin)
    plt.figure(figsize=(18,4*len(q2_bins)))
    
    for index, q2_bin in enumerate(q2_bins):
        data_sub = data.query('q2_bin == %d' % q2_bin)
        q2_value = np.unique(data_sub.q2)
    
        plt.subplot(len(q2_bins), 4, index+1)
#        plt.semilogy()
        plt.plot(data_sub.w, data_sub.xs_bosted_corr, label='Bosted')
        plt.plot(data_sub.w, data_sub.xs_keppel, label='Keppel')
        
        if xs is not None:
            xs_sub = xs.query('q2_bin == %d' % q2_bin)
            plt.errorbar(x=xs_sub.w, y=xs_sub.value*6, yerr=xs_sub.error*6, linestyle='', marker='o', color='black', label='E1-F')
        
        plt.legend(loc='best')
        plt.title('$Q^{2} = %.3f$' % q2_value, fontsize=FONT_SIZE)
        plt.grid(color='black', alpha=0.3)
#        plt.ylim([0.0, 2*np.max(data_sub.xs_keppel)])
        plt.xlabel('$W \; [GeV/c^{2}]$', fontsize=FONT_SIZE)
        plt.ylabel('$d\sigma \; [\mu b]$', fontsize=FONT_SIZE)
    plt.tight_layout()
    
def ratio_compare(data, q2bin):
    data_sub = data.query('q2_bin == %d' % q2bin)
    plt.plot(data_sub.w, data_sub.xs_bosted/data_sub.xs_keppel)
```

```python
direct_compare(data, xs)
plt.savefig('compare.pdf', dpi=500, bbox_inches='tight')
```
